# Importing Dependencies

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
#Scatter ploting 
plt.figure(figsize=(9,8))
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
```

```python
# value assigning 
x = df.iloc[:,0:1]
y = df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 10)
```

```python
# importing LinearRegression Model

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# Model Training

lr.fit(x_train, y_train)
```

**Example**

```python
# Predicting the package based on cgpa
# whith the help of example predictin the output

lr.predict(x_test.iloc[0].values.reshape(1,1))           #shape of our model is 1,1 so we reshape it using reshape function.
```

```python
x_test.iloc[0]
```

```python
# Ploting  Regression Line

plt.figure(figsize=(9,8))
plt.scatter(df['cgpa'],df['placement_exam_marks'])
plt.plot(x_train,lr.predict(x_train),color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Package (LPA)')
```

```python
# Slope calculation
m = lr.coef_   
```

```python
# Y intercept calculation
b = lr.intercept_
```

```python
print(m,b)
```

**Verification of above example with the line equation**
* **y = mx +b**
