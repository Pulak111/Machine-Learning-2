# POLYNOMIAL REGRESSION

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset 

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset
```

```python
X = dataset.iloc[:, 1:-1].values # we will want to know position, So we did not include Position
y = dataset.iloc[:, -1].values
```

# Trainig the Linear Regression model on whole DataSet

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
# We are using fit method to train the model to whole given data as we want the maximum data to trian.
lin_reg.fit(X,y) 
```

# Training the Polynomial Regression Model on whole DataSet

```python
from sklearn.preprocessing import PolynomialFeatures 

#Change the matrix of single features into matrix of new features X1,X1^2
poly_reg = PolynomialFeatures(degree = 4)

X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

# Visualising the Linear Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'green')
plt.title('Year of Experience v/s Salary')
plt.xlabel('Year of Experience')
plt.ylabel('Salary in ($)')
```

# Visualising the Polynomial Linear Regression

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title('Year of Experience v/s Salary(Polynomial Regression)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary in ($)')
```

# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

# Predicting a new result with Linear Regression

```python
lin_reg.predict([[6.5]]) # predict method always requires 2D array
```

# Predicting a new result with Polynomial Regression

```python
lin_reg2 = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
np.set_printoptions(precision = 2)
lin_reg2
```
